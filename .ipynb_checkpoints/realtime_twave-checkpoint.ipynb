# Real-Time TWave Phase Tracking

Streams an `.npz` recording through `Algo_TWave.PhaseTracker` **one sample at a time** — the same sample-by-sample loop that `Simulations.run_simulations` uses — but redraws the current window live as processing progresses, so you can watch the detector fire in context.

After the stream finishes, everything gets packaged into a `SimulationResult` so you get all the standard analysis plots (`plot_timeseries`, `plot_phase_hist`, `plot_evoked`, `plot_internals`) for free.

### What "real-time" means here

Samples are pulled from the `.npz` one at a time and fed into `tracker.update(sample)`. The tracker maintains its own internal state (`data` deque, wavelet buffer, notch filter state, …) — no look-ahead, no offline filtering. If you replaced `NpzStream` with an LSL / serial / SDK reader, everything else in this notebook would work unchanged.

Pacing is off by default (`REAL_TIME_PACING = False`) so the notebook runs as fast as Python can. Flip it on to have the loop `time.sleep` to the signal's own rate.

### Inputs

- An `.npz` with `data` (n_samples, n_channels), `fs`, and `scale_factors` — the same format `batch-processing.ipynb` uses.
- `Algo_TWave.py`, `Inhibitors.py`, and `Simulations.py` on the Python path (expected to sit next to this notebook).

## 1. Imports and configuration

```python
import time
import csv
from pathlib import Path
from collections import deque
from math import pi

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

from Simulations import (
    SimulationDataset,
    SimulationResult,
    PhaseTrackerStatus,
)
from Algo_TWave import PhaseTracker

%matplotlib inline
plt.rcParams['figure.dpi'] = 100
```

```python
# ── Input file ───────────────────────────────────────────────────────
NPZ_PATH   = Path('data/20190122-065346-004.npz')  # <-- edit me
CHANNEL_ID = 0

# ── Analysis rate ────────────────────────────────────────────────────
# TWave's wavelet (freq_limits 0.6–2 Hz, M = fs * analysis_len_s) is designed
# for a few hundred Hz, not 30 kHz raw ns6. Decimate to ~500 Hz.
ANALYSIS_RATE_HZ = 500.0

# ── Streaming controls ───────────────────────────────────────────────
MAX_DURATION_S  = 120.0   # cap processing (None → entire file)
REDRAW_EVERY_S  = 1.0     # live plot refresh rate (in signal-time)
PLOT_WINDOW_S   = 10.0    # how much recent signal to show
REAL_TIME_PACING = False  # sleep to match wall clock?

# ── TWave overrides (None → keep Algo_TWave default) ─────────────────
TARGET_PHASE = 0.0  # peak of the slow oscillation
TRACKER_OVERRIDES = {
    # 'amp_threshold_uv':      None,
    # 'amp_limit_uv':          None,
    # 'quadrature_thresh':     None,
    # 'high_low_freq_ratio':   None,
    # 'prediction_limit_s':    None,
    # 'backoff_time_s':        None,
}

# ── Ground-truth files (optional; used only by plot_timeseries) ──────
GT_SW_NPY  = None  # e.g. Path('.../ground_truth_sw.npy')
GT_IED_NPY = None  # e.g. Path('.../ground_truth_ied.npy')

# ── Output ───────────────────────────────────────────────────────────
OUTPUT_DIR = Path('../output/realtime_twave')
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
```

## 2. Streaming source

Yields one µV sample at a time from the `.npz`, decimated to `ANALYSIS_RATE_HZ`. Swap this class out for an LSL / serial / SDK reader to go truly live.

```python
class NpzStream:
    """Iterable of µV samples from an .npz, decimated to target_rate."""

    def __init__(self, path, channel_id=0, target_rate=500.0, max_duration_s=None):
        npz = np.load(str(path), allow_pickle=True)
        if 'fs' in npz:
            hw_rate = float(npz['fs'])
        elif 'sample_rate' in npz:
            hw_rate = float(npz['sample_rate'])
        else:
            raise KeyError(f'No sample rate key in {path}')

        raw = npz['data']
        if raw.ndim == 2:
            ch = min(channel_id, raw.shape[1] - 1)
            sig = raw[:, ch].astype(np.float64)
        else:
            sig = raw.astype(np.float64)

        if 'scale_factors' in npz:
            sf = npz['scale_factors']
            ch_sf = min(channel_id, len(sf) - 1)
            sig *= float(sf[ch_sf])

        ds_factor = max(1, int(round(hw_rate / target_rate)))
        sig = sig[::ds_factor]
        # Replace NaNs the same way run_simulations does.
        sig = np.nan_to_num(sig, nan=0.0)

        self.effective_rate = hw_rate / ds_factor
        if max_duration_s is not None:
            sig = sig[: int(max_duration_s * self.effective_rate)]

        self.signal    = sig
        self.hw_rate   = hw_rate
        self.ds_factor = ds_factor

        print(f'Loaded {path.name}')
        print(f'  hardware fs = {hw_rate:.1f} Hz')
        print(f'  decimation  = {ds_factor}x  →  analysis fs = {self.effective_rate:.1f} Hz')
        print(f'  duration    = {len(sig) / self.effective_rate:.1f} s  ({len(sig)} samples)')

    def __iter__(self):
        return iter(self.signal)

    def __len__(self):
        return len(self.signal)
```

## 3. Live streaming loop

Feeds samples into `tracker.update()` one at a time and redraws a three-panel view every `REDRAW_EVERY_S` of signal-time:

- **Top**: decimated signal with STIM1/STIM2 markers
- **Middle**: TWave's estimated amplitude + its `amp_threshold` and `amp_limit` lines
- **Bottom**: estimated phase (radians), with `target_phase` dashed

All per-sample `status` and `internals` values are kept so we can build a full `SimulationResult` afterwards.

```python
# ── Source ───────────────────────────────────────────────────────────
stream = NpzStream(NPZ_PATH, channel_id=CHANNEL_ID,
                   target_rate=ANALYSIS_RATE_HZ,
                   max_duration_s=MAX_DURATION_S)
fs = stream.effective_rate

# ── Tracker ──────────────────────────────────────────────────────────
tracker_kwargs = {k: v for k, v in TRACKER_OVERRIDES.items() if v is not None}
tracker = PhaseTracker(fs=int(round(fs)), target_phase=TARGET_PHASE, **tracker_kwargs)

# ── Buffers for the live view ────────────────────────────────────────
win_sp    = int(PLOT_WINDOW_S * fs)
sig_buf   = deque(maxlen=win_sp)
t_buf     = deque(maxlen=win_sp)
amp_buf   = deque(maxlen=win_sp)
phase_buf = deque(maxlen=win_sp)

# ── Full history for post-hoc SimulationResult ───────────────────────
stims_sp     = []
status_ts    = []
internals_ts = []

# ── Event counters ───────────────────────────────────────────────────
stim_times_s = []  # (t_s, which_stim, delay_sp)
inhibit_counts = {
    'BACKOFF': 0,
    'BACKOFF_ISI': 0,
    'INHIBITED_AMP': 0,
    'INHIBITED_QUADRATURE': 0,
    'INHIBITED_RATIO': 0,
    'WRONGPHASE': 0,
}
status_to_key = {
    PhaseTrackerStatus.BACKOFF: 'BACKOFF',
    PhaseTrackerStatus.BACKOFF_ISI: 'BACKOFF_ISI',
    PhaseTrackerStatus.INHIBITED_AMP: 'INHIBITED_AMP',
    PhaseTrackerStatus.INHIBITED_QUADRATURE: 'INHIBITED_QUADRATURE',
    PhaseTrackerStatus.INHIBITED_RATIO: 'INHIBITED_RATIO',
    PhaseTrackerStatus.WRONGPHASE: 'WRONGPHASE',
}

# ── Figure ───────────────────────────────────────────────────────────
fig, axes = plt.subplots(3, 1, figsize=(13, 7), sharex=True)
ax_sig, ax_amp, ax_phase = axes

def redraw(t_now):
    for ax in axes:
        ax.clear()
    t_arr = np.fromiter(t_buf, dtype=float)
    if t_arr.size == 0:
        return

    # signal + stim markers
    ax_sig.plot(t_arr, np.fromiter(sig_buf, dtype=float), color='steelblue', lw=0.8)
    t_lo = t_arr[0]
    for t_s, which, _ in stim_times_s:
        if t_s >= t_lo:
            ax_sig.axvline(t_s,
                           color='red' if which == 1 else 'darkorange',
                           ls='--', lw=1.2, alpha=0.8)
    ax_sig.set_ylabel('Signal (µV)')
    n1 = sum(1 for _, w, _ in stim_times_s if w == 1)
    n2 = sum(1 for _, w, _ in stim_times_s if w == 2)
    inh = sum(v for k, v in inhibit_counts.items() if k.startswith('INHIBITED'))
    ax_sig.set_title(
        f't = {t_now:5.1f} s   |   STIM1: {n1}   STIM2: {n2}   '
        f'inhibited: {inh}   backoff: {inhibit_counts["BACKOFF"]}'
    )

    # amplitude
    ax_amp.plot(t_arr, np.fromiter(amp_buf, dtype=float), color='seagreen', lw=0.8)
    ax_amp.axhline(tracker.amp_threshold_uv, color='grey', ls=':', lw=1,
                   label=f'amp_threshold = {tracker.amp_threshold_uv}')
    ax_amp.axhline(tracker.amp_limit_uv, color='grey', ls='--', lw=1,
                   label=f'amp_limit = {tracker.amp_limit_uv}')
    ax_amp.set_ylabel('TWave amp (µV)')
    ax_amp.legend(loc='upper right', fontsize=8)

    # phase (scatter — phase estimate is noisy when amp is low)
    ax_phase.plot(t_arr, np.fromiter(phase_buf, dtype=float),
                  color='purple', lw=0, marker='.', ms=1.5)
    ax_phase.axhline(tracker.target_phase_rad, color='green', ls='--', lw=1,
                     label=f'target = {tracker.target_phase_rad:.2f} rad')
    ax_phase.set_ylim(0, 2 * pi)
    ax_phase.set_ylabel('Phase (rad)')
    ax_phase.set_xlabel('Time (s)')
    ax_phase.legend(loc='upper right', fontsize=8)

    clear_output(wait=True)
    display(fig)

# ── Main loop ────────────────────────────────────────────────────────
redraw_every_sp = max(1, int(REDRAW_EVERY_S * fs))
dt = 1.0 / fs
wall_start = time.time()
n = -1  # so 'n + 1' on empty stream is still sane

try:
    for n, sample in enumerate(stream):
        t_now = n * dt
        result, internals = tracker.update(float(sample))

        # live-view buffers
        sig_buf.append(sample)
        t_buf.append(t_now)
        amp_buf.append(internals.get('amp', np.nan))
        phase_buf.append(internals.get('phase', np.nan))

        # full history — same shape run_simulations builds
        status_ts.append(result.status)
        internals_ts.append(internals)

        s = result.status
        if s & PhaseTrackerStatus.STIM1:
            stim_times_s.append((t_now, 1, result.delay_sp))
            stims_sp.append(n + result.delay_sp)
        if s & PhaseTrackerStatus.STIM2:
            stim_times_s.append((t_now, 2, result.delay_sp))
            stims_sp.append(n + result.delay_sp)
        for flag, key in status_to_key.items():
            if s & flag:
                inhibit_counts[key] += 1

        if (n + 1) % redraw_every_sp == 0:
            redraw(t_now)

        if REAL_TIME_PACING:
            slack = (wall_start + t_now) - time.time()
            if slack > 0:
                time.sleep(slack)

    if n >= 0:
        redraw(n * dt)  # final frame
finally:
    plt.close(fig)

# Clamp stims to valid sample indices, as run_simulations does.
stims_sp = [i for i in stims_sp if 0 <= i < len(stream.signal)]

print(f'\nDone. Processed {n + 1} samples ({(n + 1) / fs:.1f} s of signal).')
print(f'STIM1: {sum(1 for _, w, _ in stim_times_s if w == 1)}   '
      f'STIM2: {sum(1 for _, w, _ in stim_times_s if w == 2)}')
for k, v in inhibit_counts.items():
    if v:
        print(f'  {k:<22s} {v:>8d}')
```

## 4. Wrap the stream into a `SimulationResult`

Now that the history has the same shape `run_simulations` produces, we can build a `SimulationDataset` + `SimulationResult` and use the existing analysis plots.

```python
dataset = SimulationDataset(
    t=np.arange(len(stream.signal)) / fs,
    signal=stream.signal,
    fs=fs,
    name=NPZ_PATH.stem,
)

result = SimulationResult(
    Dataset=dataset,
    PhaseTracker=tracker,
    stims_sp=stims_sp,
    status_ts=status_ts,
    internals_ts=internals_ts,
)

print(f'SimulationResult built — {len(stims_sp)} stims on {len(dataset.signal)/fs:.0f} s of signal.')
```

## 5. Standard analysis plots

Each call below reuses the methods defined in `Simulations.SimulationResult`.

```python
# plot_timeseries requires ground-truth .npy files for slow-waves and IEDs.
# If you don't have them, skip this cell.
if GT_SW_NPY is not None and GT_IED_NPY is not None:
    fig = result.plot_timeseries(
        ground_truth_sw=str(GT_SW_NPY),
        ground_truth_ied=str(GT_IED_NPY),
    )
    plt.show()
else:
    print('Skipping plot_timeseries — set GT_SW_NPY and GT_IED_NPY in the config cell to enable.')
```

```python
if result.stims_sp:
    fig = result.plot_phase_hist(filt_bandpass_hz=(0.5, 4.0))
    plt.show()
else:
    print('No stims — skipping phase histogram.')
```

```python
if result.stims_sp:
    fig = result.plot_evoked(filt_bandpass_hz=(0.5, 4.0))
    plt.show()
else:
    print('No stims — skipping evoked plot.')
```

```python
fig = result.plot_internals()
plt.show()
```

## 6. Export stim log

CSV matching the batch notebook's `sample_index` convention (hardware rate).

```python
out_path = OUTPUT_DIR / f'{NPZ_PATH.stem}_stims_realtime.csv'
with open(out_path, 'w', newline='') as f:
    w = csv.writer(f)
    w.writerow(['event', 't_seconds', 'sample_index_analysis', 'sample_index_hw', 'delay_sp'])
    for t_s, which, delay_sp in stim_times_s:
        w.writerow([
            f'STIM{which}',
            f'{t_s:.4f}',
            int(round(t_s * fs)) + delay_sp,
            int(round(t_s * stream.hw_rate)) + delay_sp * stream.ds_factor,
            delay_sp,
        ])

print(f'Wrote {len(stim_times_s)} stim events → {out_path.resolve()}')
```
